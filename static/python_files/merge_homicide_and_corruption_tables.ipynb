```python
import pandas as pd
```

```python
# Read CSV Files
df_corruption = pd.read_csv("../data/Cleaned_Corruption.csv")
df_island = pd.read_csv("../data/island_info.csv")
df_homicide = pd.read_csv("../data/Homicide_Rate.csv")
```

```python
# Minor Data Adding
df_corruption = df_corruption.append({"Country": "Tonga", "CPI Score 2018": 31, "Standard error": 'NaN'}, ignore_index=True)
df_corruption
```

```python
# Merge with main islands dataframe
df_merged = df_island.merge(df_corruption, on="Country", how="left")
df_merged = df_merged[["Country", "CPI Score 2018", "Standard error"]]
```

```python
# Drop duplicates
df_merged = df_merged.sort_values("Country")
df_merged.drop_duplicates(subset ="Country", inplace = True) 
```

```python
# Update United Kingdom Value
df_homicide = df_homicide.replace({"Country": "United Kingdom of Great Britain and Northern Ireland"}, "United Kingdom")
```

```python
# merge with the main dataframe
df_merge_h = df_island.merge(df_homicide, on="Country", how="left")
df_merge_h = df_merge_h[["Country", "Homicide Rate"]]
```

```python
# Remove duplicates
df_merge_h = df_merge_h.sort_values("Country")
df_merge_h.drop_duplicates(subset ="Country", inplace = True) 
```

```python
# Merge homicde and corruption dataframes to make final dataframe
df_final = df_merge_h.merge(df_merged, on="Country", how="outer")
df_final = df_final.drop(df_final.index[len(df_final)-1])
df_final
```

```python
df_final.loc[df_final['CPI Score 2018'] > 0, 'CPI Score 2018'] = 100 - df_final.loc[df_final['CPI Score 2018'] > -1, 'CPI Score 2018'] 
```

```python
df_final = df_final.fillna(0)
df_final
```

```python
# Write that shit to a CSV
df_final.to_csv("../data/merged_countries.csv", index=False)
```
